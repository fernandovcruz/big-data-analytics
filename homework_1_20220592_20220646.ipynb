Question 1 - How many orders from STORES on June/21, in MORDOR, received any reviews?


```
%sql

SELECT count(distinct order_id)
FROM ord_sample 
WHERE order_origin = "STORE" AND MONTH(reference_date) = 6 AND city = "MORDOR" AND review_score IS NOT NULL
```

Question 2 - What is the difference in the average review_score from the first month to the last month for each order_origin? (order by difference, ascending)

```
%sql 

SELECT order_origin,
       round(AVG(CASE WHEN MONTH(reference_date) = 6 THEN review_score END) - AVG(CASE WHEN MONTH(reference_date) = 1 THEN review_score END),1) AS score_diff
FROM ord_sample
GROUP BY order_origin
ORDER BY score_diff ASC;
```

Question 4 - Considering only the last reviewed order of each customer, what is the percentage of orders that have a review_score higher than 4 for each city? (order by percentage, ascending) 

```
%sql

WITH last_order_review AS (
  SELECT account_id, MAX(review_created_at) AS last_review_date
  FROM ord_sample
  WHERE review_score IS NOT NULL
  GROUP BY account_id
),
last_order AS (
  SELECT o.*
  FROM ord_sample o
  JOIN last_order_review r
  ON o.account_id = r.account_id AND o.review_created_at = r.last_review_date
),
city_review_count AS (
  SELECT city, COUNT(*) AS total_orders, SUM(CASE WHEN review_score > 4 THEN 1 ELSE 0 END) AS reviews_abv_4
  FROM last_order
  GROUP BY city
)
SELECT city, round(reviews_abv_4 * 100.0 / total_orders,1) AS percentage
FROM city_review_count
ORDER BY percentage ASC;
```

Question 5 - What is the average number of view_checkout events per month? (answer should have 6 lines, date in format YYYY-MM-01)

```
%sql 

SELECT DATE_FORMAT(reference_date, "yyyy-MM-01") AS month,
       round(AVG(CAST(view_checkout AS INT)),1) AS avg_view_checkout
FROM sess_sample
GROUP BY month
ORDER BY month;
```

Question 6 - What is the average distance between the merchant and customer for each order origin?

Question 8 - Which city had the lowest number of new account_ids in the last month compared to the first month?

```
%sql

--This version of the question 8, which gave us as result "GONDOR" is the one we interpreted as correct, since we considered for june only the accounts which did their first order in that month. It is not our submitted response.

SELECT city, count(distinct case when month(account_first_date) = 1 then account_id end) as new_accounts_january, 
             count(distinct case when month(account_first_date) = 6 then account_id end) as new_accounts_june, 
             count(distinct case when month(account_first_date) = 6 then account_id end) - count(distinct case when month(account_first_date) = 1 then account_id end) as diff
FROM(
  SELECT account_id, city, account_first_date
  FROM (
    SELECT account_id, city, reference_date,
           MIN(reference_date) OVER (PARTITION BY account_id) AS account_first_date
    FROM ord_sample
  ) 
  WHERE MONTH(account_first_date) = 1 OR MONTH(account_first_date) = 6
)
group by city
order by diff asc
```

```
%sql 

--We later decided to re-do our query following the professor's response in moodle's Discussion Forum "Maybe it is simple. Accounts last month minus account first month." even if it seemed like it wasn't considering the first month of appearence of "new accounts" in june.

SELECT city, count(distinct case when month(reference_date) = 1 then account_id end) as new_accounts_january, 
             count(distinct case when month(reference_date) = 6 then account_id end) as new_accounts_june, 
             count(distinct case when month(reference_date) = 6 then account_id end) - count(distinct case when month(reference_date) = 1 then account_id end) as diff
FROM ord_sample
GROUP BY city
ORDER BY diff asc
```

Question 9 - What is the cumulative sum of order_total for each month over time? (order by date, ascending)

```
%sql

SELECT 
    DATE_FORMAT(reference_date, 'yyyy-MM-01') AS month,
    round(SUM(SUM(order_total)) OVER (ORDER BY DATE_FORMAT(reference_date, 'yyyy-MM-01')),1) AS cumulative_sum
FROM ord_sample
GROUP BY DATE_FORMAT(reference_date, 'yyyy-MM-01')
ORDER BY month ASC;



```

Question 10 - What is the average difference in days between orders made by the same customer?

```
%sql 

SELECT round(AVG(DATEDIFF(reference_date, prev_order_date)),1) AS avg_days_diff
FROM (
  SELECT account_id, reference_date, 
         LAG(reference_date) OVER (PARTITION BY account_id ORDER BY reference_date) AS prev_order_date
  FROM ord_sample
) t
WHERE prev_order_date IS NOT NULL;


```
